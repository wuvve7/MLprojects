```python
# Task 3: Data Preparation

import pandas as pd
from sklearn.preprocessing import StandardScaler

# 1. Load the dataset
url = "https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv"
df = pd.read_csv(url)

# 2. Display first 5 rows
print("Preview of the dataset:")
print(df.head())

# 3. Dataset info
print("\nDataset Information:")
print(df.info())

# 4. Check for missing values
print("\nMissing Values in Each Column:")
print(df.isnull().sum())

# 5. Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")
if duplicates > 0:
    df = df.drop_duplicates()
    print("Duplicate rows removed.")

# 6. Summary statistics
print("\nSummary Statistics:")
print(df.describe())

# 7. Check for unexpected values in categorical columns
print("\nValue Counts for Categorical Variables:")
print("Sex:\n", df['sex'].value_counts())
print("\nSmoker:\n", df['smoker'].value_counts())
print("\nRegion:\n", df['region'].value_counts())

# 8. One-Hot Encoding for categorical variables
df_encoded = pd.get_dummies(df, columns=['sex', 'smoker', 'region'], drop_first=True)

# 9. Normalize numerical features using StandardScaler
scaler = StandardScaler()
numerical_cols = ['age', 'bmi', 'children']
df_encoded[numerical_cols] = scaler.fit_transform(df_encoded[numerical_cols])

# 10. Final prepared dataset preview
print("\nPrepared Dataset (after cleaning, encoding and scaling):")
print(df_encoded.head())
```
